```python
import google.generativeai as genai
from google.colab import userdata
import torch
from diffusers import StableDiffusionPipeline
import random
import sys
import gradio as gr
```

## Preparando Gemini

```python
userdata.get('apikey_gemini')
gemini_api_secret_name = 'apikey_gemini'

try:
    GOOGLE_API_KEY = userdata.get(gemini_api_secret_name)
    genai.configure(api_key=GOOGLE_API_KEY)
except userdata.SecretNotFoundError as e:
    print(f'''Secret not found\n\nThis expects you to create a secret named {gemini_api_secret_name} in Colab\n\nVisit https://makersuite.google.com/app/apikey to create an API key\n\nStore that in the secrets section on the left side of the notebook (key icon)\n\nName the secret {gemini_api_secret_name}''')
    raise e
except userdata.NotebookAccessError as e:
    print(f'''You need to grant this notebook access to the {gemini_api_secret_name} secret in order for the notebook to access Gemini on your behalf.''')
    raise e
except Exception as e:
    # unknown error
    print(f"There was an unknown error. Ensure you have a secret {gemini_api_secret_name} stored in Colab and it's a valid key from https://makersuite.google.com/app/apikey")
    raise e

model = genai.GenerativeModel('gemini-pro')
```

```python
modelos_disponiveis = [
    "redstonehero/cetusmix_v4",
    "stablediffusionapi/edge-of-realism",
    "DGSpitzer/Cyberpunk-Anime-Diffusion",
    "stablediffusionapi/disney-pixar-cartoon"
]
for i in modelos_disponiveis:
  md = i
  pipe = StableDiffusionPipeline.from_pretrained(f"{md}", torch_dtype=torch.float16)
  pipe = pipe.to("cuda")

```

## Declarando Funções

```python
def gerar_script(product, mainAdvantage):
  pd = product
  adv = mainAdvantage
  prompt_gen = f"Create a funny commercial for {pd} complete with eye-catching logos and slogans, showcasing its main advantage: {adv}. The scene must have a prompt to generate a image maximum of 40 word, maintaining a fast-paced and unpredictable storyline. Remember the prompt has to be really happy and do not forget the logo"
  response = model.generate_content(prompt_gen)
  prompt_gen = response.text
  seed = random.randint(0, sys.maxsize)
  return prompt_gen
```

```python
def comedias(text, adv, modelo ,neg_prpt):
  model = modelo.split(" ")
  md = model[2]

  pipe = StableDiffusionPipeline.from_pretrained(f"{md}", torch_dtype=torch.float16)

  pipe = pipe.to("cuda")
  pipe.safety_checker = None
  prpt = gerar_script(text, adv)
  prompt = prpt
  h=640
  w=640
  steps=30
  guidance=7.5
  neg = neg_prpt

  image = pipe(prompt, height=h, width=w, num_inference_steps=steps, guidance_scale=guidance, negative_prompt=neg).images[0]
  return image
```

# Application

```python
# Lista de modelos disponíveis
modelos_disponiveis = [
    "anime - redstonehero/cetusmix_v4",
    "realismo - stablediffusionapi/edge-of-realism",
    "realismo2 - DGSpitzer/Cyberpunk-Anime-Diffusion",
    "disney_pixar - stablediffusionapi/disney-pixar-cartoon"
]

# Interface Gradio
inputs = [
    gr.Textbox(label="Nome do Produto"),
    gr.Textbox(label="Vantagem do produto"),
    gr.Dropdown(choices=modelos_disponiveis, label="Modelo"),
    gr.Textbox(label="Negative Prompt")
]

output = gr.Image(label="Imagem gerada")

```

```python
gr.Interface(fn=comedias, inputs=inputs, outputs=output, title="ComédIAds" ).launch(debug = True, share = True)
```
